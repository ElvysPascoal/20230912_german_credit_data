```python
# Dispersão entre 'Credit amount' e 'Duration'
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Credit amount', y='Duration')
plt.title('Relação entre Valor do Crédito e Duração')
plt.xlabel('Valor do Crédito')
plt.ylabel('Duração')
plt.show()
```

```python
# Boxplot para variáveis numéricas
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[['Age', 'Credit amount', 'Duration']])
plt.title('Boxplot para Variáveis Numéricas')
plt.xlabel('Variável')
plt.ylabel('Valor')
plt.show()
```

A imagem abaixo mostra um gráfico de caixa que representa a relação entre “Saving accounts” e “Credit amount”. Uma possível interpretação da imagem é que ela compara o valor do crédito solicitado por pessoas com diferentes níveis de poupança. O gráfico de caixa mostra o valor mínimo, o primeiro quartil, a mediana, o terceiro quartil e o valor máximo de cada categoria de poupança. O gráfico de caixa também pode mostrar a presença de valores discrepantes, que são pontos que estão muito distantes da média ou da mediana.

A partir do gráfico, podemos observar algumas características das categorias:

A categoria “little” tem a maior mediana de valor do crédito, o que pode indicar que as pessoas com pouca poupança tendem a solicitar créditos mais altos, possivelmente por necessidade ou falta de planejamento financeiro.
A categoria “moderate” tem a menor mediana de valor do crédito, o que pode indicar que as pessoas com uma poupança moderada tendem a solicitar créditos mais baixos, possivelmente por cautela ou responsabilidade financeira.
A categoria “rich” tem a maior variação de valor do crédito, o que pode indicar que as pessoas com uma poupança alta têm mais flexibilidade e diversidade na hora de solicitar créditos, possivelmente por oportunidade ou preferência pessoal.
A categoria “quite rich” tem uma variação intermediária de valor do crédito, o que pode indicar que as pessoas com uma poupança bastante alta têm um equilíbrio entre necessidade e oportunidade na hora de solicitar créditos.

```python
# Boxplot para 'Saving accounts' por 'Credit amount'
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Saving accounts', y='Credit amount')
plt.title('Relação entre "Saving accounts" e "Credit amount"')
plt.xlabel('Saving accounts')
plt.ylabel('Credit amount')
plt.xticks(rotation=45)
plt.show()
```
